The dataset has the following information:

* Pregnancies: Number of times pregnant
* Glucose: Plasma glucose concentration over 2 hours in an oral glucose tolerance test
* BloodPressure: Diastolic blood pressure (mm Hg)
* SkinThickness: Triceps skin fold thickness (mm)
* Insulin: 2-Hour serum insulin (mu U/ml)
* BMI: Body mass index (weight in kg/(height in m)^2)
* DiabetesPedigreeFunction: A function that scores the likelihood of diabetes based on family history.
* Age: Age in years
* Outcome: Class variable (0: a person is not diabetic or 1: a person is diabetic)

```python
#quitamos lo que no usaremos
Pokemon=Pokemon.drop(['percentage_male','type','generation'],axis=1)
Pokemon
```

```python
# Entrem mas gordo y mas grande mas lento puede ser, para compenzarlo subiremos su velocidad base dependiendo de su altura y peso
Pokemon2=Pokemon.drop(['is_legendary'],axis=1) #quitamos la ultima columna para comodidad al unir una nueva columna
Friccion = (Pokemon.height_m + Pokemon.weight_kg)/10 #hacemos una lista que suma la altura y el peso y lo divide por 10
Friccion = pd.DataFrame(Friccion, columns = ['Friccion']) # lo hacemos un dataframe
Pokemon2=pd.concat([Pokemon2,Friccion], axis=1) #unimos el nuevo dataframe
Pokemon=pd.concat([Pokemon2,Pokemon.is_legendary], axis=1) #por ultimo unimos la columna que quitamos por comodidad
```

```python
#Velocidades resultantes luego de la equilibracion 
sp_attackdef = (Pokemon.sp_attack + Pokemon.Friccion)
sp_defensedef = (Pokemon.sp_defense +  Pokemon.Friccion)
speeddef = (Pokemon.speed +  Pokemon.Friccion)  #hemos sumado las velocidades con lo que llamamos friccion
sp_attackdef = pd.DataFrame(sp_attackdef, columns = ['sp_attackdef'])
sp_defensedef = pd.DataFrame(sp_defensedef, columns = ['sp_defensedef'])
speeddef = pd.DataFrame(speeddef, columns = ['speeddef']) #los hicimos dataframes
Pokemon2=Pokemon.drop(['is_legendary'],axis=1) 
Pokemon2=pd.concat([Pokemon2,sp_attackdef,sp_defensedef,speeddef], axis=1)
Pokemon=pd.concat([Pokemon2,Pokemon.is_legendary], axis=1) #Por ultimo unimos todo en nuestro dataframe
#Pokemon = Pokemon.drop(['height_m','sp_attack','sp_defense','speed','weight_kg'],axis=1)
Pokemon
```

```python
# El ataque, defensa y vida es lo mas importante seguido de la velocidad, asi mismo la velocidad de ataque y defensa tabien tienen peso 
PoderPuchamon = (Pokemon.attack*5 + Pokemon.defense*5 + Pokemon.hp*5 + Pokemon.sp_attackdef*3 + Pokemon.sp_defensedef*3 + Pokemon.speeddef*4 )/20
PoderPuchamon = pd.DataFrame(PoderPuchamon, columns = ['PoderPuchamon'])
Pokemon2=Pokemon.drop(['is_legendary'],axis=1)
Pokemon2=pd.concat([Pokemon2,PoderPuchamon], axis=1)
Pokemon=pd.concat([Pokemon2,Pokemon.is_legendary], axis=1)
Pokemon.to_csv('Pokemon.csv')
#haremos 
```
